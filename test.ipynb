```python
yearly_data = df.groupby("Year")["Dowry_Death_Cases"].sum()

plt.figure()
plt.plot(yearly_data.index, yearly_data.values, marker='o')
plt.title("Year-wise Dowry Death Cases in India")
plt.xlabel("Year")
plt.ylabel("Number of Dowry Death Cases")
plt.show()
```

```python
state_data = df.groupby("State")["Dowry_Death_Cases"].sum()

plt.figure()
plt.bar(state_data.index, state_data.values)
plt.title("State-wise Dowry Death Cases")
plt.xlabel("State")
plt.ylabel("Dowry Death Cases")
plt.xticks(rotation=45)
plt.show()
```

```python
total_dowry = df["Dowry_Death_Cases"].sum()
total_crimes = df["Total_Crimes_Against_Women"].sum()

plt.figure()
plt.pie(
    [total_dowry, total_crimes - total_dowry],
    labels=["Dowry Deaths", "Other Crimes"],
    autopct='%1.1f%%'
)
plt.title("Dowry Crimes Share Among Crimes Against Women")
plt.show()
```

```python
area_data = df.groupby("Area_Type")["Dowry_Death_Cases"].sum()

plt.figure()
plt.bar(area_data.index, area_data.values)
plt.title("Urban vs Rural Dowry Death Cases")
plt.xlabel("Area Type")
plt.ylabel("Dowry Death Cases")
plt.show()
```

```python
conviction_trend = df.groupby("Year")["Conviction_Rate_Percent"].mean()

plt.figure()
plt.plot(conviction_trend.index, conviction_trend.values, marker='o')
plt.title("Average Conviction Rate Over Years")
plt.xlabel("Year")
plt.ylabel("Conviction Rate (%)")
plt.show()
```
